```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
plt.plot(x,np.sin(x), '-', label = 'sine')
plt.plot(x, np.cos(x), '--', label = 'cosine')

plt.xlim([0, 10])
plt.ylim([-1, 1])
plt.xlabel('x')
plt.ylabel('amplitude')
plt.title('My sines')
plt.grid()
plt.legend();
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x), '-')
ax.plot(x, np.cos(x), '--');
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x), '-', label = 'sine')
ax.plot(x, np.cos(x), '--', label = 'cosine')
ax.set(xlim =[0, 10], ylim = [-1, 1],
        xlabel = 'x', ylabel = 'amplitude',
        title = 'My sines');
ax.grid()
ax.legend();
```

```python
fig.savefig('sines.pdf')
```

```python
data = pd.read_csv('auto-mpg.csv', na_values = '?')
```

```python
data.plot()
```

```python
data['mpg'].plot.hist();
```

```python
data.origin.value_counts()
```

```python
axs = data.hist(column = 'mpg', by = 'origin')
```

```python
axs = data.hist(column='mpg', by='origin')
axs[1].set(title='1', ylim=[0, 55])
axs[2].set(title='2', ylim=[0, 55])
axs[3].set(title='3', ylim=[0, 55]);
```

```python
data.plot.scatter('mpg', 'horsepower', c='origin', colormap='winter');
```

```python
fix, ax = plt.subplots()
data.plot.scatter('mpg','horsepower', c= 'origin', colormap = 'viridis', ax = ax);
```

```python
colors = {1: 'tab:blue', 2: 'tab:orange', 3: 'tab:green'}
fig, ax = plt.subplots()
for key, group in data.groupby(by='origin'):
    group.plot.scatter('mpg', 'horsepower', c = colors[key], label = key, ax = ax);
```
